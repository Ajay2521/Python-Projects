# CREATE YOUR STICKY NOTES

## PROGRAM CONCEPT:

In this I have divided the program file into 4 different functional parts, namely:
    
    - note_creation(num)
    - print_notes(name_given,notes_given)
    - save_notes(name_given,notes_given)
    - main()

Lets see one by one:

### note_creation(num):

    In this the function is created by using one parameter called num. Thus by using that num parameter the for loop is iterated for the range of num to get num number of values from user. 
    
    In simple this function is used to get notes one by one from user till user required value get satisfied.
    
### print_notes(name_given,notes_given):

    In this by passing out the name and notes, by using the enumerated for loop the notes were printed one by one.
    
### save_notes(name_given,notes_given):

    In this by passing out the name and notes, a file with the given name is created with ".txt" extension with mode as write mode. Then by using enumerated for loop with file write() the notes will be writen into the file one by one.
    
### main()

    This is the main function of the program which is used to call every other function which is created before. Initially it get user name and then number of notes user needed. then by passing number of nodes to note_creation() the user notes will be created and by passing the user name and user notes into print_notes() and save_notes() the notes will be printed and saved by using the user name. 


## PROGRAM SOURCE CODE

```python
%%writefile sticky_note.py

'''
python program for creating our own sticky notes
'''

def note_creation(num):
    '''
    function for creating notes from user.
    '''
    notes = []
    for i in range(1,num+1):
        note = input("\nEnter your Note - {} :".format(i))
        notes.append(note)
    return notes

def print_notes(name_given,notes_given):
    '''
    function to print the sticky notes.
    '''
    print()
    print(name_given.capitalize(),"- here is your Sticky Notes")
    for i,note_given in enumerate(notes_given,1):
        print("\nNote",i,"-",note_given)

def save_notes(name_given,notes_given):
    '''
    function to save the yours stickynote as txt file.
    '''
    file_name = "%s_Sticky_Notes.txt" %name_given.capitalize()
    file = open(file_name, "w")
    file.write("\n{} - here is your Sticky Notes".format(name_given.capitalize()))
    for i,note_given in enumerate(notes_given,1):
        file.write("\n\nNote {} - {}".format((i),note_given))
    file.close()
    print("\nFILE HAS BEEN SAVED SUCCESSFULLY...")

def main():
    '''
    function which is used to call other functions.
    '''
    print("\nCREATE YOUR OWN STICKY NOTES...")
    user_name = input("\nEnter your Name - ")
    notes_num = int(input("\nEnter number of notes needed - "))
    user_notes = note_creation(notes_num)
    print_notes(user_name,user_notes)
    print()
    save_notes(user_name,user_notes)
```
